```python
def limpiar_columnas_constantes(df):
    columnas_constantes = []
    
    for columna in df.columns:
        if df[columna].nunique() == 1:
            columnas_constantes.append(columna)
    
    df = df.drop(columnas_constantes, axis=1)
    
    return df
```

```python
df_1 = limpiar_columnas_constantes(df)
print(df_1)
```
